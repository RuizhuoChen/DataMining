```python
import pandas as pd
import math
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

#read order_data.csv and create a DataFrame with that content
data = pd.read_csv(r"order_data.csv",delimiter=" ",header=None)
#preprocessing: change to one hot encoding so as to be able to use apriori from mlxtend
d=data.values.tolist()

#removing nan values
for i in range(len(d)):
    j=0
    while(True):
        if (type(d[i][j])==float and math.isnan(d[i][j])) :
            del d[i][j]
            j-=1
        j+=1
        if (j>len(d[i])-1):
            break
```

```python
te = TransactionEncoder()
te_ary = te.fit(d).transform(d)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
#computing frequent itemsets and association rules
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
a=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.9)

#visualizing association rules results
a[["antecedents","consequents","support","confidence"]]
```

```python
data = pd.read_csv(r"mammographic_masses.csv",delimiter=",",header=0,usecols=['Age','Severity'])

#preprocessing: change to one hot encoding so as to be able to use apriori from mlxtend
d=data.values.tolist()
#removing nan values
for i in range(0,len(d)):
    j=0
    while(True):
        if (type(d[i][j])==float and math.isnan(d[i][j])) :
            del d[i][j]
            j-=1
        j+=1
        if (j>len(d[i])-1):
            break
#adding attributes
for i in range(len(d)):
    for j in range (len(d[i])):
        d[i][j]=data.columns[j] + "=" +str(d[i][j])
te = TransactionEncoder()
te_ary = te.fit(d).transform(d)

df = pd.DataFrame(te_ary, columns=te.columns_)
#computing frequent itemsets and association rules
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
a=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.01)

#visualizing association rules results
a[["antecedents","consequents","support","confidence"]]
```

```python
data = pd.read_csv(r"mammographic_masses.csv",delimiter=",",header=0)

#preprocessing: change to one hot encoding so as to be able to use apriori from mlxtend
d=data.values.tolist()
#removing ? values
i=0
while(True):
    if (d[i][1]=='?') :
        del d[i]
        i-=1
    i+=1
    if (i>len(d)-1):
        break
        
#adding attributes
for i in range(len(d)):
    for j in range (len(d[i])):
        if j == 1 and int(d[i][j])<40:
            d[i][j] = 'Young'
        elif j == 1 and int(d[i][j])>=40:
            d[i][j] = 'Old'
        d[i][j]=data.columns[j] + "=" +str(d[i][j])
        
te = TransactionEncoder()
te_ary = te.fit(d).transform(d)

df = pd.DataFrame(te_ary, columns=te.columns_)
#computing frequent itemsets and association rules
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)

a=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.9)

#visualizing association rules results
a[["antecedents","consequents","support","confidence"]]
```

```python
for i in range(len(a)):
    if ("Age=Old" in a["antecedents"][i] and "Severity=1" in a["consequents"][i] ):
        print (a["antecedents"][i],a["consequents"][i],a["support"][i],a["confidence"][i] )
```
