```python
from sklearn.cluster import KMeans
```

```python
import numpy as np
import pandas as pd

df_data = pd.read_csv('clustering_data.csv')
select_col= df_data.columns[1:]
data = np.array(df_data[select_col])
data
```

```python
kmeans_default = KMeans(init='random', n_clusters=8, max_iter=300, n_init=10).fit(data)
kmeans_init = KMeans(init='k-means++', n_clusters=8, max_iter=300, n_init=10).fit(data)
kmeans_n_init = KMeans(init='random', n_clusters=8, max_iter=300, n_init=100).fit(data)
```
